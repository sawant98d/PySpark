```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Dataframe Filter').getOrCreate()
```

```
df_pyspark = spark.read.csv('/FileStore/tables/test_data-3.csv', header=True, inferSchema=True)
```

```
df_pyspark.show()
```

#### Filter Operation

```
### Lets select a records of emp whose salary is less than or equal to 20000
df_sal = df_pyspark.filter('salary<=20000') # here salary/column name is case insensetive 
df_sal.show()
```

```
df_sal = df_pyspark.filter('salary<=20000').select(['Name','Age'])
df_sal.show()
```

```
# Below is single condition in filter function
# We can also write a condition in below mentioned way. Its alternative for above one
df_sal = df_pyspark.filter(df_pyspark['Salary']<=20000) 
df_sal.show()
```

```
# How to write multiple condition in filter function, below is demostration..
df_sal = df_pyspark.filter((df_pyspark['Salary']<=20000) & (df_pyspark['Salary']>=15000)) # here salary/column name is case insensetive 
df_sal.show()
```

```
df_sal = df_pyspark.filter((df_pyspark['Salary']<=20000) | (df_pyspark['Salary']>=15000)) # here salary/column name is case insensetive 
df_sal.show()
```

```
df_sal = df_pyspark.filter(~(df_pyspark['Salary']<=20000))
df_sal.show()
```
