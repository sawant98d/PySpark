#### Tutorial 5- Pyspark With Python-GroupBy And Aggregate Functions
#### PySpark GroupBy and Aggregate Functions

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('GroupBynAgregate').getOrCreate()
```

```
df_pyspark = spark.read.csv('/FileStore/tables/test_data-4.csv', header=True, inferSchema=True)
df_pyspark.show()
```

```
df_pyspark.printSchema()
```

```
## Grouped by Department to find maximum salary for that particular department

df_agg = df_pyspark.groupBy('Department').sum() # Groupby followed by agg or sum or ,..functions
df_agg.show()
```

```
df_pyspark.groupBy('Department').max().show() # Here we found the maximun/ highest salary for each department
```

```
# Lets find mean of the grouped departments
df_agg = df_pyspark.groupBy('Department').mean() 
df_agg.show()
```

```
# Lets find number of employees are working for each departments
df_agg = df_pyspark.groupBy('Department').count() 
df_agg.show()
```

```
df_pyspark.agg({'Salary':'sum'}).show() # here we have calculated aggreated total salary..
```
