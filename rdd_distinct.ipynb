```
# Demostration of distinct()

from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Distinct Demo")
sc = SparkContext.getOrCreate(conf=conf)
rdd = sc.textFile("/FileStore/tables/sample.txt")
rdd.collect()  # Reading simple RDD
```

```
flatRDD = rdd.flatMap(lambda x:x.split())
flatRDD.collect() #reading flat RDD
```

```
distRDD = flatRDD.distinct()
distRDD.collect() # reading distinct element from RDD using distinct()
```
