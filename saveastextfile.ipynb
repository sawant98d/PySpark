```
# saveastextfile is an action which saves the spark RDD to text file
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('count action')
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/sample_words.txt')
rdd.collect()
```

```
rdd2 = rdd.flatMap(lambda x:x.split(' '))
rdd2.collect()
rdd2.getNumPartitions()
```

```
rdd2.saveAsTextFile('/FileStore/tables/output/sample_words_output.txt') #save file in distributed fashion save data in two partitions
```
