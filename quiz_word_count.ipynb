```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('quiz word count')
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/word_count_quiz.txt')
```

```
#For the quiz youâ€™ll be using this input file
#Quiz Statement
"""
this mango company 
cat mango ant animal laptop
chair switch mango am charger cover
animalany mango ant laptop laptop
this 

Read this file in the RDD
Write a transformation flow that will return the word count of each word present in the file as (key, value) pair
"""
```

```
flat_rdd = rdd.flatMap(lambda x:x.split(' '))
#flat_rdd.collect()
map_rdd = flat_rdd.map(lambda x:(x,1))
#map_rdd.collect()
grp_by_key_rdd = map_rdd.reduceByKey(lambda x,y:x+y)
grp_by_key_rdd.collect()
```
