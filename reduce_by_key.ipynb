```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName('reducebykey')
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/sample_words.txt')
rdd.collect()
```

```
rdd2 = rdd.flatMap(lambda x:x.split(' '))
rdd2.collect()
```

```
rdd3 = rdd2.map(lambda x:(x,len(x)))
rdd3.collect()
```

```
rdd4 = rdd3.reduceByKey(lambda x,y:x+y) # here in reducebykey we need to provide the way how to reduce the key, here we have passed lambda fun
rdd4.collect()

#VIMP : here groupByKey works like,
#suppose there is a word 'this' which occures 4 times in rdd so ('this',[4,6,3,2])
#and for reducebykey it will add 4+6=10 10+3=13 13+2=15 so return 15 from reducebykey fun
```
