```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName('groupbykey')
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/sample_words.txt')
rdd.collect()
```

```
rdd2 = rdd.flatMap(lambda x:x.split(' ')).map(lambda x:(x,len(x)))
rdd2.collect()
```

```
rdd3 = rdd2.groupByKey()
rdd3.collect()
```

```
rdd3.mapValues(list).collect() # we want the group of the values in a list that is why we have provided list as parameter to mapValues method
```
